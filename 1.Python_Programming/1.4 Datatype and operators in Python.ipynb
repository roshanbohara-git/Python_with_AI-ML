# Datatypes
- Data types are a classification of data which tell the compiler or interpreter how the programmer intends to use the data
- They determine the type of opertations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data

## Importance:
- Data types ensure that data is stored in an efficient way.
- They help in performing correct operations on data.
- Proper use of data types can prevent errors and bugs in the program.


```python
# Basic Data Types
# Interger Example

age = 35
type(age)
```

```python
# Float point data type
height = 5.8
print(height)
print(type(height))
```

```python
# String data type example
name = "Roshan"
print(name)
print(type(name))
```

```python
## boolean datatype
is_student = True
print(is_student)
print(type(is_student))
```

```python
a = 10
b = 10
type(a==b)
```

```python
## common errors we get while working with different daatypes 
result = "Hello"+5
```

```python
result = "Hello"+str(5)
print(result)
```
