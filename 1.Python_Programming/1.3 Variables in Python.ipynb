# Variables
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a varibale. 

**Variable** is a name given to memory location. A variable name can be pointer to the stored value(object). A variable can contain int, float, string, list, tuple, set, bool, predefined_variable etc. Variables are of two types: Local variable and Global Variable.

```python
a = 100
```

### Naming Conventions
1. Variables names should be descriptive and in small letter.
2. They must start with a letter or an '_" and contain letter, numbers and underscores
3. A variable can start from letter and underscore but it cannot start with a number.
4. Variables names are case sensitive
5. A variable can only contain alpha-numberic characters(a-Z, 0-9) and underscore(_).

```python
# Declaring and asigning variables

age = 22
height = 5.8
name = "Roshan"
is_student = True

# printing the variables
print("Age :", age)
print("Height:", height)
print("Name:", name)
```

```python
# Invalid variable names
1age = 30
first-name = "Roshan"
@name = "Roshan"
```
