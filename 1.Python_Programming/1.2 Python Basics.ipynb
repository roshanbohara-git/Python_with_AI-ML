## Basic Syntax Rules in Python:
1. Case Sensitivity
2. Indentation
3. Line Conitinuation
4. Multiple Statements on a single line


```python
# 1. Case Sensitivity
name = "Roshan"
Name = "Bohara"
print(name)
print(Name)
```

**Indentation**: Indentation in Python is used to define the structure and hirerarchy of the code. Unlike many other programming languages that use braces {} to delimt blocks of code. Python uses indentation to determine the grouping of statements. This means that all the statements within a block must be indented at the same level.

```python
# 2. Indentation: Python uses indentation to define blocks of code. Consistent use of spaces(commomnly 4) or a tab is required.

age = 32
age1 = 20
if age>30:
    print(f"Your age is {age}")  # Indentation

print(age1)
```

```python
# 3. Line Continuation : Use a backslash(\) to continue a statement to the next line
total = 1 +2 +2 +5 +10 +\
20 + 50
print(total)
```

```python
# 4. Multiple Statements on a single line
x = 5; y = 10; z = x + y
print(z)
```
